# Missing Values

```python
## Here we will check the % of NaN values present in each features
## 1 - step make the list of featues which has missing values

features_with_NaN = [features for features in dataset.columns if dataset[features].isnull().sum() > 1]

## 2- step print the features name and the % of missing values
for features in features_with_NaN:
    print(features +'(missing value): % ', np.round(dataset[features].isnull().mean(), 4))
```

# OR

```python
#the above line of code can be viewed in a more simplified way as:
## code to get the all the columns having NaN values
features_with_NaN = [] 
for features in dataset.columns:
    if dataset[features].isnull().sum() > 1:
        features_with_NaN.append(features)

#printing the features with NaN %(NULL / Total)
for features in features_with_NaN:
    print(features, np.round(dataset[features].isnull().mean(), 4), '% missing values')
```

# Since there are many missing values, we need to find the relationship between the missing values and the  Sale Price

Let' plot some diagram for this relationship

```python
# Iterating through all the Fetures having Null values
for feature in features_with_NaN:
    data = dataset.copy()
    
    #let's make a variable that indicates 1 if the observation was missing or 0 if not.
    data[feature] = np.where(data[feature].isnull(),1,0)
    
    #let's caculate the mean SalePrice where the information is missing or present
    # Here groupby features helps taking the two columns at a time.
    data.groupby(feature)['SalePrice'].median().plot.bar(color = ['gray','green'])
    plt.title(feature)
    plt.show()
```

For most of the missing values featuers, we find that the NaN values(1) have greater median values than the normal data points.
That is the relation between the missing values and the dependent variables is clearly visible. So we need to replace these NaN values with something meaningful. 
We will do this part in Feature Engineering section.

# Numerical Values

```python
# visualizing the numerical features
dataset[numerical_features].head()
```

Temporal Variables(Eg: Datetime variables)
From the dataset we have 4 Years variables. We have exact information from the datetime variables like no of years or no of days. One example in this specific scenario can be difference in years between the years the house was built and the year the house sold out. We will be performing this analysis in the Feature Engineering.

```python
#list of variables that contain year information
year_feature = []
for feature in numerical_features:
    if 'Yr' in feature or  'Year' in feature:
        year_feature.append(feature)
```

```python
year_feature
```

```python
#let's explore the content of these year variables.
for feature in year_feature:
    #print(feature)
    print(np.sort(dataset[feature].unique()))
```

```python
# let's analyze the Temporal Datetime variables
## we will check whether there is any realation b/w year the house is sold and the sales price
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year sold')
plt.ylabel('Median House Price')
plt.title('House price vs  Year sold')
```

Here, we see that the Median Sales price is decreasing with year
but this is not a good data trend. price should be increasing.

```python
# here we will compare the difference of [other years]-YrSold with the salePrice(so that we get the life span)
for feature in year_feature:
    if feature != 'YrSold':
        data = dataset.copy()
        #capturing the diff of Year var and Year house was sold: lifetime
        data[feature] = data['YrSold']-data[feature]
        
        plt.scatter(data[feature], data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

From the above scatter plots, we find that the the density decreases with years.
i.e., in the initial year of construction or initial year of re-modification the price was high 
and it decreased over time.
Here we are tracking for the acutual size or time-duration of any process.
And it is obvious to get such plots.

```python
# Numerical variables are usually of 2 types
## 1. continuous variables and Discrete Variables
discrete_feature = []
for feature in numerical_features:
    if len(dataset[feature].unique()) < 25 and feature not in year_feature+['Id']:
        discrete_feature.append(feature)
```

# Continuous Variable

```python
#logic for separating continous variable
continuous_feature = []
for feature in numerical_features:
    data = dataset.copy()
    if feature not in discrete_feature and feature not in year_feature + ['Id']:
        continuous_feature.append(feature)
    
```

```python
print('count of continuous features: ', len(continuous_feature))
```

# Relation Between Continuous Features and Sale Price

```python
# Plotting histogram for each continuous feature v/s sale Price
for feature in continuous_feature:
    data = dataset.copy()
    data[feature].hist(bins = 30)
    plt.xlabel(feature)
    plt.ylabel('Sale Price')
    plt.title(feature)
    plt.show()
```

for continuous variables we plot the distribution of the data,
From the above plots, we find some of them follows gaussing/normal distribution but 
some of them are skewed so we need to convert them to standard normal distribution 
(may be using log normal ).

# Exploratory Data Analysis Part 2


Whenever the data is not in the form of gaussian distribution(i.e., its skewed), it is good to transform the data.

```python
# We will be using Logarithmic transformation
#logarithmic transformation helps linearlize/remove skewness/ thus reducing the variance.

for feature in continuous_feature:
    if 0 in data[feature].unique(): #log(0) !!! not defined
        pass
    else:
        
        data = dataset.copy()
        data[feature] = np.log(data[feature])
        data['SalePrice'] = np.log(data['SalePrice'])
        plt.scatter(data[feature], data['SalePrice'])
        plt.title(feature)
        plt.xlabel(feature)
        plt.ylabel('Sale Price')
        plt.show()
```

```python
# before applying log transformation, the plots were skewed in one side
# We see positive correlation in the above scatter plots, i.e., processed to be used in regression
```

# Outliers

```python
# outliers are the points which are either very high or very low from the rest of the datapoints.
for feature in continuous_feature:
    data = dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:  
        data[feature] = np.log(data[feature])
        data.boxplot(column = feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()

    
```

```python
#box plot is generaly used for continuous variables
#Its ranges are Q1, Q2, Q3, IQR = (Q3-Q1), upper whisker, and lower whisker= Q1/Q3 +- 1.5* IQR
# data out of the upper and lower whiskers are outliers.
```
